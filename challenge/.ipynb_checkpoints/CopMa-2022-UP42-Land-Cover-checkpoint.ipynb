```python
# Import python modules
import json
import up42
import leafmap
import geopandas as gpd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Authenticate and initialize project
# https://sdk.up42.com/authentication/
up42.authenticate(cfg_file="project_conf_file.json")

project = up42.initialize_project()
```

```python
# Create a new workflow
workflow = project.create_workflow(name="CopMa 2022: Land Cover using Pléiades", 
                                   use_existing = True)
```

```python
# Before adding the blocks to your new workflow, check the names of available blocks
block_list=up42.get_blocks(basic=False)
#print(block_list)

for block in range(len(block_list)):
  name_list=json.dumps(block_list[block]["displayName"], 
                       ensure_ascii=False)
  print(name_list)
```

```python
# Selection of the datablock and the processing block we want to use to start our project
input_tasks = ["Pléiades Reflectance (Download)"]

workflow.add_workflow_tasks(input_tasks)
```

```python
# Get compatible blocks
workflow.get_compatible_blocks()
```

```python
# Selection of the datablock for Land Cover Classifier
# https://up42.com/marketplace/blocks/processing/landcover
input_tasks = ["Pléiades Reflectance (Download)",
              "pansharpen",
              "Land Cover Classifier for Pléiades/SPOT "]

workflow.add_workflow_tasks(input_tasks)
```

```python
# Define your Area of Interest
#aoi = up42.read_vector_file("{/computer/path/where/aoi/is/stored}.geojson")
aoi = up42.read_vector_file(Path.cwd().joinpath("aoi_tree.geojson"))
```

### Before I start, I would like to check my area of interest (AOI) to confirm that my AOI will have and that its location is correct.

```python
# Making Leafmap using AOI
gdf = gpd.read_file(Path.cwd().joinpath("aoi_tree.geojson"))
m = leafmap.Map()
m.add_gdf(gdf, layer_name="Saint Malo")
m
```

## We now return to the use of the UP42 SDK.

```python
# Adjust the JSON parameters like in the job configuration window
input_parameters = workflow.construct_parameters(geometry = aoi, 
                                                 geometry_operation = "bbox",
                                                 start_date = "2020-06-01",
                                                 end_date = "2022-05-01",
                                                 limit = 1)

# Dont forget to adjust the parameters.
input_parameters["oneatlas-pleiades-fullscene:1"].update({"max_cloud_cover":20})
input_parameters["pansharpen:1"].update({"clip_to_aoi":True})
input_parameters["pansharpen:1"].update({"bbox": [-2.0129871368408203, 48.64050901118123, 
                                                 -1.9929027557373047, 48.65043321555078]})

input_parameters
```

```python
# Before running the job, estimate the credits that will be deducted from your account
workflow.estimate_job(input_parameters)
```

```python
# Run the actual job.
job = workflow.run_job(input_parameters, track_status=True)
```

```python
# Downlod & Visualize results.
path = job.download_results()[0]
# Plot of quicklook: the latest and cloudless image
img = mpimg.imread(path)
imgplot = plt.imshow(img)
plt.show()
```
